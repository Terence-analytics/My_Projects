```python
# LOADING DATA  
import pandas as pd

df = pd.read_excel(r'C:\Users\Me\Downloads\ict 2022 mentorship\data analyst bootcamp\Tutorials\Customer Call List.xlsx')
df.head(5)  # display first 5 rows
```

```python
# inspect data
df
''' 
    drop not useful column
    data contains a duplicate row
    first name column its ok
    last name column contains special characters
    standardise phone number column
    split address into 3 columns [Street_adress,city,zipcode]
    standardize paying customer and do not contact columns

'''
```

```python
# Drop not useful column
df.drop(columns='Not_Useful_Column', inplace=True)
df
```

```python
#remove duplicate rows
df.drop_duplicates(inplace=True)
df
```

```python
# remove special characters on last name column
df['Last_Name'] = df['Last_Name'].str.replace('[^a-zA-Z]','', regex=True)
df
```

```python
# Standardize phone number column
# change everything to str and remove special characters
df['Phone_Number']= df['Phone_Number'].apply(lambda x : str(x))
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)
df
```

```python
# Standardize phone number column
# change the format numbers to xxx-xxx-xxxx
df['Phone_Number']= df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-'+ x[6:])
df['Phone_Number']= df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']= df['Phone_Number'].str.replace('Na--','')
df
```

```python
# remove customer with no address
df= df[df['Address'] !='N/a']
df
```

```python
# split address
df[['Street_Address','City','Zip_Code']] = df['Address'].str.split(',',n=2, expand=True)
df
```

```python
# drop address column
df.drop(columns= 'Address', inplace=True)
df
```

```python
# Standardize Paying customer 
df['Paying Customer']= df['Paying Customer'].str.replace('No','N')
df['Paying Customer']= df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']= df['Paying Customer'].str.replace('N','No')
df['Paying Customer']= df['Paying Customer'].str.replace('Y','Yes')
df['Paying Customer'] = df['Paying Customer'].str.replace('No/a','')
df
```

```python
# Standardize DO_NOT_CONTACT 
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('N','No')
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('Y','Yes')
df['Do_Not_Contact']= df['Do_Not_Contact'].apply(lambda x: str(x))
df['Do_Not_Contact']= df['Do_Not_Contact'].str.replace('nan','Yes')
df
```

```python
df = df.reset_index(drop=True)
df
```

# BASIC ANALYSIS

```python
# How many customers are paying vs not paying?
paying_customers = len(df[df['Paying Customer']=='Yes'])
not_paying_customers = len(df[df['Paying Customer']=='No'])
print("Total paying customers: ", paying_customers ,"\nTotal not paying customers: ",not_paying_customers )
```

```python
# How many customers can be contacted?
ContactableCustomers = len(df[df['Do_Not_Contact']=='Yes'])
print("customers that can be contacted: ", ContactableCustomers)
```
