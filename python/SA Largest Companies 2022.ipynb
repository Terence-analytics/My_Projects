```python
#WEB SCRAPING FROM : https://en.wikipedia.org/wiki/List_of_largest_companies_in_South_Africa
```

```python
# IMPORT LIBRARIES
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Connect to website and get the content(html)
URL = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_South_Africa"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)","Accept-Language": "en-US,en;q=0.9",}
page = requests.get(URL, headers= headers)
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

```python
# Extract data
# Table
table = soup.find('table', class_='wikitable sortable')
table
```

```python
# Column names
columns_html = table.find_all('th') # html
columns = [title.text.strip() for title in columns_html]
print(columns)
```

```python
# put columns in dataFrame using pandas
df = pd.DataFrame(columns= columns)
df
```

```python
#extract data

rows = table.find_all('tr')[1:] # rows from index 1
for row in rows:
    row_data = row.find_all('td') # table data by row
    individual_row_data = [data.text.strip() for data in row_data] # a list of row x data
    length = len(df)   # length of our data frame
    df.loc[length]= individual_row_data # insert row data into our dataFrame

df
```
